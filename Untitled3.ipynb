```python
import kagglehub
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
import os
```

```python
# Download latest version
path = kagglehub.dataset_download("uciml/iris")
```

```python
print("Path to dataset files:", path)
```

```python
base = pd.read_csv(os.path.join(path, "Iris.csv"), keep_default_na=False)
base.head()
```

```python
base = base.drop(columns=['Id'])
base.head()
```

```python
y = base['Species'].values
x = base.drop(columns=['Species']).values
x
```

```python
labelencoder = LabelEncoder()
for i in range(x.shape[1]):
  if x[:,i].dtype == 'object':
    x[:,i] = labelencoder.fit_transform(x[:,i])
x
```

```python
x_treinamento, x_test, y_treinamento, y_test = train_test_split(x, y, test_size=0.3, random_state=12)
```

```python
modelo = RandomForestClassifier(random_state=1, max_depth=20, max_leaf_nodes=12, n_estimators=500)
modelo.fit(x_treinamento, y_treinamento)
```

```python
tree_index = 0
tree_to_visualize = modelo.estimators_[tree_index]
plt.figure(figsize=(20, 20))
plot_tree(tree_to_visualize, filled=True, feature_names=base.columns[:-1], class_names=True, rounded=True)
plt.show
```

```python
previsoes = modelo.predict(x_test)
previsoes
```

```python
accuracy = accuracy_score(y_test, previsoes)
precision = precision_score(y_test, previsoes, average='weighted')
recall = recall_score(y_test, previsoes, average='weighted')
f1 = f1_score(y_test, previsoes, average='weighted')
print(f'accuracy: {accuracy}')
print(f'precision: {precision}')
print(f'recall: {recall}')
print(f'f1: {f1}')
```

```python
report = classification_report(y_test, previsoes)
print(report)
```
